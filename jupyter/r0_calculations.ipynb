```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from matplotlib import rc
from scipy import stats as st
from scipy.special import logsumexp
from scipy.signal import savgol_filter

rc('text', usetex=True)
rc('text.latex', preamble=r'\usepackage{cmbright}')
rc('font', **{'family': 'sans-serif', 'sans-serif': ['Helvetica']})

%matplotlib inline

# This enables SVG graphics inline. 
%config InlineBackend.figure_formats = {'png', 'retina'}

# JB's favorite Seaborn settings for notebooks
rc = {'lines.linewidth': 2, 
      'axes.labelsize': 18, 
      'axes.titlesize': 18, 
      'axes.facecolor': 'DFDFE5'}
sns.set_context('notebook', rc=rc)
sns.set_style("dark")

mpl.rcParams['xtick.labelsize'] = 16 
mpl.rcParams['ytick.labelsize'] = 16 
mpl.rcParams['legend.fontsize'] = 14
```

```python
# load into a dataframe:
# df = pd.read_csv('~/my_repos/covid-19-data/us-states.csv')
# df.date = pd.to_datetime(df.date)
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv'
df = pd.read_csv(url, usecols=[0, 1, 3, 4], parse_dates=['date'], squeeze=True)
```

```python
def sir(c0, gamma, r):
    """Returns expected number of cases at next time point:"""
    return c0 * np.exp(gamma * (r - 1))
```

```python
# we want to calculate lambda through time:
def p_lambda_given_k(k, l):
    """
    Params:
    -------
    l:    float. burst size of infections
    k:    int. number of infections
    """
    return st.poisson.pmf(k, l)
```

```python
ny = df[(df.state == 'New York') & (df.cases > 20)].copy()
ny['newCases'] = ny.cases.diff().rolling(window=5, win_type='gaussian', center=True).mean(std=2).round()
ny['newDeaths'] = ny.deaths.diff().rolling(window=5, win_type='gaussian', center=True).mean(std=2).round()
```

```python
def bayes_prob(t, r, cases, prev, p_prev=1):
    l = prev * np.exp(gamma * ( r - 1))
    # probability function:
    if t > 0:
        prior = st.norm(loc=r, scale=0.25).pdf(r[:, None])
        prior /= prior.sum()
    else:
        prior = st.norm(loc=4, scale=.01).pdf(r[:, None])
        prior /= prior.sum()


    p = st.poisson.logpmf(cases, l)
    p = np.exp(p - logsumexp(p)) * p_prev @ prior
    p /= p.sum()
    
    return p


gamma = 1 / 10
colors = cm.viridis(np.linspace(0, 1, len(ny)))

r = np.linspace(.5, 9, 1000)

t = 0
maxR = np.repeat(-1., len(ny))
maxp = 0
prev = 1
p_prev = 1
for date, g in ny.groupby('date'):
    if np.isnan(g.newCases.values[0]):
        t += 1
        continue

    p = bayes_prob(t, r, g.newCases.values[0], prev, p_prev)
    p_prev = p
    prev = g.newCases.unique()[0]
    
    plt.plot(r, p, color=colors[t])

    maxR[t] = r[np.where(p == p.max())]
    t += 1

plt.axvline(1)
plt.xlabel(r'$R_0$')
```

```python
def r_calc(d, col='newCases'):
    t = 0
    r = np.linspace(.5, 9, 1000)
    maxR = np.repeat(-1., len(d))
    maxp = 0
    prev = 1
    p_prev = 1
    for date, g in d.groupby('date'):
        if np.isnan(g[col].values[0]):
            t += 1
            continue

        p = bayes_prob(t, r, g[col].values[0], prev, p_prev)
        prev = g[col].unique()[0]

        if np.isnan(p).all():
            maxR[t] = -1
        else:
            p_prev = p
            maxR[t] = r[np.where(p == p.max())]

        t += 1
    return maxR
```

```python
states = ['New York', 'California', 'Massachusetts', 'Washington']
fig, ax = plt.subplots(ncols=len(states), nrows=2, figsize=(20, 8), sharey=True, sharex=True)

i = 0
for n, g in df[df.cases > 10].groupby('state'):
    if n not in states:
        continue
    gp = g.copy()
    gp['newCases'] = gp.cases.diff().rolling(window=5, win_type='gaussian', center=True).mean(std=2).round()
    gp['newDeaths'] = gp.deaths.diff().rolling(window=5, win_type='gaussian', center=True).mean(std=2).round()

    maxR = r_calc(gp)
    maxRd = r_calc(gp, col='newDeaths')
    
    x = (gp.date - gp.date.min()) / dt.timedelta(days=1)
    ax[0, i].plot(x[maxR > 1],
             maxR[maxR > 1],
             'o', color='red')
    ax[0, i].plot(x[(maxR <= 1) & (maxR > 0)],
             maxR[(maxR <= 1) & (maxR > 0)],
             'd', color='blue')

    ax[1, i].plot(x[maxRd > 1],
             maxRd[maxRd > 1],
             'o', color='red')
    ax[1, i].plot(x[(maxRd <= 1) & (maxRd > 0)],
             maxRd[(maxRd <= 1) & (maxRd > 0)],
             'd', color='blue')
    ax[0, i].set_title(n)
    i += 1

for ai in ax:
    for aij in ai:
        aij.axhline(1, color='black', zorder=0, ls='--')
        aij.set_ylim(0, 8)

ax[0, 0].set_ylabel('R0 based on cases')
ax[1, 0].set_ylabel('R0 based on deaths')

_ = plt.figtext(.5, .045, 'Days since 1 death', fontsize=25, ha='center')
```

```python
total = df.groupby('date')[['cases', 'deaths']].sum()
total.reset_index(inplace=True)

total['newCases'] = total.cases.diff().rolling(window=5, win_type='gaussian', center=True).mean(std=2).round()
total['newDeaths'] = total.deaths.diff().rolling(window=5, win_type='gaussian', center=True).mean(std=2).round()

total = total[total.newDeaths > 1]

maxR = r_calc(total)
maxRd = r_calc(total, col='newDeaths')

fig, ax = plt.subplots(nrows=2, figsize=(10, 7), sharey=True, sharex=True)

x = (total.date - total.date.min()) / dt.timedelta(days=1)
ax[0].plot(x[maxR > 1],
         maxR[maxR > 1],
         'o', color='red')
ax[0].plot(x[(maxR <= 1) & (maxR > 0)],
         maxR[(maxR <= 1) & (maxR > 0)],
         'd', color='blue')

ax[1].plot(x[maxRd > 1],
         maxRd[maxRd > 1],
         'o', color='red')
ax[1].plot(x[(maxRd <= 1) & (maxRd > 0)],
         maxRd[(maxRd <= 1) & (maxRd > 0)],
         'd', color='blue')

for ai in ax:
    ai.axhline(1, color='black', zorder=0, ls='--')
    ai.set_ylim(0, 6)

ax[0].set_ylabel('R0 based on cases')
ax[1].set_ylabel('R0 based on deaths')

_ = plt.figtext(.5, 0, 'Days since 1 death', fontsize=25, ha='center')
```
